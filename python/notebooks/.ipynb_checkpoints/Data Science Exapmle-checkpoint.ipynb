```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np


X = np.array(c[['passenger_count', 'trip_distance']])
Y = np.array(c[['fare_amount']])

regr = linear_model.LinearRegression()

regr.fit(X, Y)

y_pred = regr.predict(X)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(Y, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(Y, y_pred))
```

```python
import json
import sys
from sqlalchemy import create_engine
import pandas as pd


def getRAPData(en, sqlStatement):
    engine = create_engine(en)
    a=pd.read_sql_query(sqlStatement,con=engine)
    b=json.dumps(a.values[0][0])
    c = pd.DataFrame.from_dict(dict(json.loads(b))["data"])
    return c

    
```

```python
engine = 'postgresql://drehc098h01isk.cl35knpvok5p.us-east-2.rds.amazonaws.com:6534/dre_dev?user=stageuser&password=goWest123!'

SQL = """Select stage.svc_data_viewer('{"source_id":236,"columns":["fare_amount","passenger_count","payment_type","trip_distance","tpep_dropoff_datetime","tpep_pickup_datetime"],
"page":1,"page_size":100,"order_by":"tpep_pickup_datetime","where":"passenger_count=''1''"}'::json)"""

data = getRAPData(engine, SQL)
data.to_csv("RAPDATA.csv")
data.head()
```

```python
from sklearn.externals import joblib
import numpy as np
import datetime
import pandas as pd

now = datetime.datetime.now()

modelName = "taxi hello world"
modelLocation = '../trained_models/TaxiFareRegression.pkl'

engine = 'postgresql://drehc098h01isk.cl35knpvok5p.us-east-2.rds.amazonaws.com:6534/dre_dev?user=stageuser&password=goWest123!'

SQL = """Select stage.svc_data_viewer('{"source_id":236,"columns":["fare_amount","passenger_count","payment_type","trip_distance","tpep_dropoff_datetime","tpep_pickup_datetime"],
"page":1,"page_size":100,"order_by":"tpep_pickup_datetime","where":"passenger_count=''1''"}'::json)"""

predFields = ['passenger_count', 'trip_distance']

predModel = joblib.load(modelLocation)

data = getRAPData(engine, SQL)

X = np.array(data[predFields])

predictions = predModel.predict(X)
predictions = predictions.tolist()
predictions = [x[0] for x in predictions]
timestamp = [now.strftime("%Y-%m-%d %H:%M")] * len(predictions)
model = [modelName] * len(predictions)
#dummy ids for test
ids = list(range(0,len(predictions)))
classes = ["Taxi Fare"] * len(predictions)

finalDFDict = {"predictions": predictions, "timestamp": timestamp, "model_name": model, "id": ids, "predicted_class": classes}
finalData = pd.DataFrame.from_dict(finalDFDict)
finalData.to_sql("stage.predictions", engine, if_exists="append")

```
